19. create a new column called Age_Group:
    - Child : age<18
    - Adult :  10 to 60
    - Senior : age>60
      Then count passengers in each category

```python
conditions=[
    df['age'] < 18,
    ( df['age'] >= 18) &  (df['age'] <= 60),
     df['age'] > 60
]
categories= ["Child","Adult","Senior"]
df["Age_Group"] = np.select (conditions,categories, default="Unknown")
print(df['Age_Group'].value_counts())
```

20. Find survival percentage for each passengers class

```python
survival_per=df.groupby('pclass')['survived'].mean()*100
print(survival_per)
```

```python
# formatting it
res = survival_per.reset_index()
res.columns=['Passenger Class','Survival Percentage']
print(res)
```
